```python
### making a copy of the dataset 
data1 =data.copy()
```

```python
### checking for null values
data1.isnull().sum()
```

```python
### dealing with null values
for x in data1.columns:
    data1[x].fillna(data1[x].mode()[0],inplace=True)
```

```python
### checking for null values
data1.isnull().sum()
```

## Mapping Delivery_Mode

```python
data1.Delivery_Mode=data1.Delivery_Mode.map({'Mode-1':3,'Mode-2':2,'Mode-3':4,'Mode-4':1,'Mode-5':5})
```

## Mapping Sales_Agent

```python
## mapping Sales_Agent
data1.Sales_Agent=data1.Sales_Agent.map({'Sales-Agent-1':1,'Sales-Agent-2':2,'Sales-Agent-3':3,'Sales-Agent-4':4,'Sales-Agent-5':5,'Sales-Agent-6':6,'Sales-Agent-7':7,'Sales-Agent-8':8,'Sales-Agent-9':9,'Sales-Agent-10':10,'Sales-Agent-11':11,'Sales-Agent-12':12})
```
